```python
import os
import json
import itertools
import torch
import soundfile as sf
import whisperx
from datasets import load_dataset

def seconds_to_srt_timestamp(seconds: float) -> str:
    """
    Convert seconds (float) to SRT timestamp format: HH:MM:SS,mmm
    """
    hours = int(seconds // 3600)
    minutes = int((seconds % 3600) // 60)
    secs = int(seconds % 60)
    millis = int(round((seconds - int(seconds)) * 1000))
    return f"{hours:02d}:{minutes:02d}:{secs:02d},{millis:03d}"

# Load and stream the Hindi dataset from Hugging Face.
dataset_stream = load_dataset("ai4bharat/indicvoices_r", "Hindi", split="train", streaming=True)
samples = list(itertools.islice(dataset_stream, 200))

# Create directories for audio, metadata, and srt files.
os.makedirs("audio", exist_ok=True)
os.makedirs("hindi_samples/metadata", exist_ok=True)
os.makedirs("srt_files", exist_ok=True)

# Determine the device to use for WhisperX (GPU if available, else CPU)
device = "cuda" if torch.cuda.is_available() else "cpu"

# Load the alignment model for Hindi using the language code "hi"
align_model, metadata = whisperx.load_align_model(language_code="hi", device=device)
```

```python


# Process each sample: save audio, metadata, run alignment, and generate SRT files.
for i, sample in enumerate(samples):
    # Save the audio to a WAV file.
    audio_array = sample["audio"]["array"]
    sr = sample["audio"]["sampling_rate"]
    audio_path = f"audio/{i+1}.wav"
    sf.write(audio_path, audio_array, sr)

    # Save metadata (removing the audio array to minimize file size).
    sample_meta = sample.copy()
    sample_meta.pop("audio")
    meta_path = f"hindi_samples/metadata/sample_{i+1}.json"
    with open(meta_path, "w", encoding="utf-8") as f:
        json.dump(sample_meta, f, ensure_ascii=False, indent=2)

    # Create the transcription segment from the sample.
    transcription_dict = sample
    segments = [{
        "start": 0.0,
        "end": transcription_dict["duration"],
        "text": transcription_dict["text"]
    }]

    # Run alignment using the segments and the audio file.
    aligned = whisperx.align(segments, align_model, metadata, audio_path, device)
    word_segments = aligned.get("word_segments", [])

    # Build the SRT file content.
    srt_lines = []
    for idx, word_info in enumerate(word_segments, start=1):
        start_timestamp = seconds_to_srt_timestamp(word_info["start"])
        end_timestamp = seconds_to_srt_timestamp(word_info["end"])
        # Each SRT block: index, time range, and subtitle text (word)
        srt_block = f"{idx}\n{start_timestamp} --> {end_timestamp}\n{word_info['word']}\n"
        srt_lines.append(srt_block)

    srt_text = "\n".join(srt_lines)
    srt_path = f"srt_files/{i+1}.srt"
    with open(srt_path, "w", encoding="utf-8") as f:
        f.write(srt_text)
    print(f"Sample {i+1}: SRT saved at {srt_path}")
```
